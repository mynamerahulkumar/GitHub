```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

## Data load and transformation

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
label = pd.DataFrame(list(map(lambda x : iris.target_names[x],iris.target)),columns=['Species'])
```

```python
iris = pd.concat([data,label],axis=1);iris.head()
```

```python
iris.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
```

## distplot()

```python
sns.set_style('whitegrid') #default : 'white'

sns.distplot(iris.SepalLengthCm) 
```

## hist

```python
sns.distplot(iris.SepalLengthCm,norm_hist=False,  fit=None, kde=False) 
```

```python
#set size
plt.figure(figsize=(12,6))
#plot
plt.subplot(2,2,1)
sns.distplot(iris.SepalLengthCm)
plt.subplot(2,2,2)
sns.distplot(iris.SepalWidthCm)
plt.subplot(2,2,3)
sns.distplot(iris.PetalLengthCm)
plt.subplot(2,2,4)
sns.distplot(iris.PetalWidthCm)
```

## boxplot()

```python
sns.boxplot(data = iris.drop(['Species'],axis=1))
```

## countplot()
1 category feature

```python
sns.countplot(x='Species',data=iris)
```

## pairplot()
1 continuous vs 1 continuous

```python
sns.pairplot(iris[['SepalLengthCm','SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']], hue='Species')
```

## lmplot()
1 continuous vs  1 continuous

```python
sns.lmplot(x='SepalLengthCm',y='PetalLengthCm',data=iris)
```

```python
sns.lmplot(x='SepalLengthCm',y='PetalLengthCm',data=iris, fit_reg=False, hue='Species') #scatter
```

## barplot()
1 continuous vs 1 category

```python
sns.barplot(x='Species',y='SepalLengthCm',data=iris) #mean
```

## correlation
1 continuous vs 1 continuous

```python
cmap = sns.diverging_palette(200, 0, as_cmap=True)
sns.heatmap(iris.corr(),cmap='coolwarm',annot=True, fmt='.2f')
```
