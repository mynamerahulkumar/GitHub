## stacking

```python
import numpy as np
x1=np.arange(10,20,1)
y1=np.arange(20,30,1)
print(x1.size)
print(y1.size)
print(x1)
print(y1)
```

```python
z1=np.hstack([x1,y1])  # horizontally concating two stack
z1
```

```python
z1=np.vstack([x1,y1])
z1
```

```python
x2=np.arange(4).reshape(2,2)
y2=np.arange(4).reshape(2,2)
print(x2)
print(y2)
```

```python
np.hstack([x2,y2])
```

```python
np.vstack([x2,y2])
```

```python
# z dimension stack 
z3=np.dstack([x2,y2])
print(z3.shape)
```

```python
np.column_stack([x2,y2])
```

```python
#Spliting array
x1=np.arange(10,22,2) # start =0 ,end=<20 ,step=2
np.split(x1,2) # both the sides number should be same -even counts
```
