###
- Filter records for (Sachin R Tendulkar,virendar sehwag,Rahul Dravid]
- Create a heatmap,where rows will have player names ,columns will have versus,values ,equal average

```python
odi=pd.read_csv('odi-batting.csv')
```

```python
players=['Sachin R Tendulkar','Virender Sehwag','Rahul Dravid']
```

```python
odi_subset=odi[odi['Player'].isin (players)]
```

```python
summary_log=odi_subset.groupby(['Player','Versus']).agg({'Runs':'mean'})
summary_log.head()
```

```python
summary_wide=summary_log.unstack(level=1) #level=1 player will be in rows,0- player will be in column
summary_wide
```

```python
sns.heatmap(summary_wide,cmap='Blues')
```

```python
#index='player'
#columns='versus'
#values='Runs'
#aggfun='mean'
#when you know the index use pivot table 
summary_pivot=pd.pivot_table(odi_subset,
                            index='Player',
                            columns='Versus',
                            values='Runs',
                            aggfunc='mean')
summary_pivot
```
