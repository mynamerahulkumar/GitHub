```python
player_summary=odi.groupby(['Player']).agg({'Runs':'sum'})
#player_summary.loc['Sachin R Tendulkar','Runs']
player_summary.sort_values(['Runs'],ascending=False).head(10)
```

```python
# CALCULATE PLAYER WISE TOTAL CENTURY and identiy top 10
player_summary=odi[['Player','Runs']]
player_summary=player_summary[player_summary['Runs']>99]
player_summary.groupby(['Player']).agg({'Runs':'count'}).sort_values('Runs',ascending=False).head(10)
```

# Apply Methods


```python
odi['Runs']+odi['ScoreRate']
```

```python
odi['is_century']=odi['Runs'].apply(lambda run:1 if run >99 else 0)
odi['is_duck']=odi['Runs'].apply(lambda run:1 if run==0 else 0)
odi['is_fifty']=odi['Runs'].apply(lambda run:1 if run >49 and run <100 else 0)
odi['missed_century']=odi['Runs'].apply(lambda run:1 if run >90 and run <100 else 0)
odi.head()
```

```python
player_summary=odi.groupby(['Player']).agg({
    'Runs':'sum',
    'is_century':'sum',
    'is_duck':'sum',
    'missed_century':'sum',
    'ScoreRate':'mean',
    'Country':'count'
})
player_summary=player_summary.rename(columns={'Country':'No. of Matches'})
player_summary.sort_values(['Runs'],ascending=False).head(10)
```

```python
#Calculate yearwise total run for Sachin R Tendulkar
odi['year']=odi['MatchDate'].apply(lambda d:d[-4:])
sachin_rows=odi[odi['Player']=='Sachin R Tendulkar']
sachin_years=sachin_rows.groupby(['year']).agg({'Runs':'sum'})
sachin_rows
```

```python
#type of variable to join :DF+DF
# Row or Columns join?:Columnwise
#common key:
```

```python
#rahul_rows=odi[odi['Player'].str.contains('Rahul')]
rahul_rows=odi[odi['Player']=='Rahul Dravid']
rahul_years=rahul_rows.groupby(['year']).agg({'Runs':'sum'})
rahul_years
```

```python
df_new=pd.concat([sachin_years,rahul_years],axis=1,sort=False,join='inner')
df_new
```

```python
df_new=pd.concat([sachin_years,rahul_years],axis=1,sort=False,join='outer')
df_new.columns=['Runs_Sachin','Runs_Rahul']
df_new
```

```python
df_new=pd.concat([sachin_years.rename(columns={'Runs':'Runs_sachin'}),
                 rahul_years.rename(columns={'Runs':'Runs_Rahul'})],
                axis=1,
                sort=False,
                join='outer')
df_new
```

# Joining


```python
df_new=sachin_years.join(rahul_years,lsuffix='_sachin',rsuffix='_rahul')
df_new.head(10)
```

```python
#rahul_rows=odi[odi['Player'].str.contains('Rahul')]
virat_rows=odi[odi['Player']=='Virat Kohli']
virat_years=virat_rows.groupby(['year']).agg({'Runs':'sum'})
virat_years
```

```python
df_new=sachin_years.join([rahul_years.rename(columns={'Runs':'rahul'}),
                         virat_years.rename(columns={'Runs':'virat'})],
                         sort=False,
                         how='outer')
df_new
```
