```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

# Data Visulization

```python
odi=pd.read_csv('odi-batting.csv')
```

```python
top_players=odi['Player'].value_counts().head(10)
top_players.index
```

```python
xvalues=top_players.index
yvalues=top_players.values
plt.figure(figsize=(10,3))
plt.bar(xvalues,yvalues)
plt.xticks(rotation=45)

plt.xlabel('Player')
plt.ylabel('Total Matches')
plt.title('Top 10 Players by total No. of Matches')
```

```python
top_players.plot.bar()
```

```python
odi['Player'].value_counts().head(10).plot.barh(color='steelblue')
```

# linechart

```python
odi['year']=odi['MatchDate'].str[-4:]
odi['year'].unique()
sachin_rows=odi[odi['Player'].str.contains('Sachin')]
years_runs=sachin_rows.groupby(['year'])['Runs'].sum() # this returns a series and agg returns a dataframe
years_runs.plot.line()
```

```python
plt.figure() # connecting matplot lib with pandas
years_runs.cumsum().plot.line() # adding the run year wise
plt.legend(['Total Runs: %d' % years_runs.sum()])
```

```python
hr=pd.read_csv('https://bit.ly/2zwLZ5I')
hr.head()
```

```python
hr['MonthlyIncome'].describe()
```

```python
hr['MonthlyIncome'].sort_values().iloc[735]
```

```python

hr['MonthlyIncome'].plot.box()
```

```python
hr.boxplot('MonthlyIncome',by='JobRole',figsize=(15,5)) # sales representative salary difference is less
plt.show()# Sales representative dots are outlier (could be)
```

```python
sales=hr[hr['JobRole']=='Sales Representative']
quantiles=sales['MonthlyIncome'].quantile([0.25,0.75])
iqr=quantiles.loc[0.75]-quantiles.loc[0.25]
lw=quantiles.loc[0.25]-1.5*iqr
uw=quantiles.loc[0.75]+1.5*iqr
lw,uw
```

```python
sales[sales['MonthlyIncome']<lw]
```

```python
sales[sales['MonthlyIncome']>uw]
```
