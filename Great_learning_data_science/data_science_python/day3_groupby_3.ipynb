
# duplicates

```python
ds=pd.Series(['a','a','c','d','a'])
ds.duplicated() # first time 'a' not duplicated second time 'a ' is duplicated 
```

```python
ds.drop_duplicates()
```

```python
df=pd.DataFrame({
    'a':[1,2,3,1],
    'b':[1,2,4,1]
})
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
# count no of duplicates
df.shape[0]-df.drop_duplicates().shape[0]
```
