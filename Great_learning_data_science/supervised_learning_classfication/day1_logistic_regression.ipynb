## Logistics regression-Statistics Approach


```python
import statsmodels.api as sm
y=df1['TenYearCHD']
X=df1.drop(columns='TenYearCHD')
```

```python
X=sm.add_constant(X) 
```

```python
model=sm.Logit(y,X)
result=model.fit()
result.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
pd.DataFrame({'vif':vif},index=X.columns)[1:]
```

```python
## backward elimination to drop insignificant variable one by one
```

```python
cols=list(X.columns)
pmax=1
while(len(cols)>0):
    pvalues=sm.Logit(y,X).fit().pvalues(1:)
    pmax=pvalues.max()
    pid=pvalues.idmax()
    if pmax>0.05:
        cols=cols.remove(pid)
    
    
```
